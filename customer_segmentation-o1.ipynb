dropping the irrelevant columns

```python
columns_to_drop=["Preferred Payment\n Method"]
cust_data.drop(columns=columns_to_drop, inplace=True)
cust_data.head()
```

Renaming the columns


```python
cust_data.rename(columns={
    'Purchase \nAmount (USD)': 'Purchase Amount',
    'Review \nRating' : 'Rating',
    'Subscription \nStatus': 'Subscription',
    'Payment \nMethod' : 'Payment Method',
    'Shipping \nType' : 'Shipping Type',
    'Discount \nApplied' : 'Discount',
    'Promo Code\n Used' : 'Promo_Code_Used',
    'Previous\n Purchases' : 'Old Purchases',
    'Frequency of\n Purchases' : 'Frequency of Purchases',
    'Previous Purchased \nAmount (USD)' : 'Old Purchase Amount',
    'Average \nOrder Value (USD)' : 'Average Order Value (USD)', 
    'Total Sales \nRevenue Per Customer (USD)' : 'Total Sales',
    'Conversion\n Rate %' : 'Conversion Rate'
}, inplace = True)

cust_data.columns
```

```python
cust_data['Purchase Amount']=cust_data['Purchase Amount'].str.replace('$','').astype(float)
cust_data['Old Purchase Amount']=cust_data['Old Purchase Amount'].str.replace('$','').astype(float)
cust_data['Average Order Value (USD)']=cust_data['Average Order Value (USD)'].str.replace('$','').astype(float)
cust_data['Total Sales']=cust_data['Total Sales'].str.replace('$','').astype(float)
cust_data.head()
```

```python
cust_data['Click-through rate (CTR)']=cust_data['Click-through rate (CTR)'].str.replace('%','').astype(float)
cust_data['Conversion Rate']=cust_data['Conversion Rate'].str.replace('%','').astype(float)
cust_data.head()
```

```python
import matplotlib.pyplot as plt

# Filter for male purchases
male_purchases = cust_data[cust_data['Gender'] == 'Male']

# Group by category and sum the purchase amounts
category_totals = male_purchases.groupby('Category')['Purchase Amount'].sum()

# Plotting
plt.figure(figsize=(10, 6))
category_totals.plot(kind='bar', color='skyblue')
plt.title('Total Male Purchases by Category')
plt.xlabel('Category')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
category_totals.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8))
plt.title('Male Purchases Distribution by Category')
plt.ylabel('')
plt.show()
```

```python

Female_purchases = cust_data[cust_data['Gender'] == 'Female']

# Group by category and sum the purchase amounts
category_totals = Female_purchases.groupby('Category')['Purchase Amount'].sum()

# Plotting
plt.figure(figsize=(10, 6))
category_totals.plot(kind='bar', color='pink')
plt.title('Total Female Purchases by Category')
plt.xlabel('Category')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by age group and category, and sum the purchase amounts
age_category_totals = cust_data.groupby(['Age', 'Category'])['Purchase Amount'].sum().unstack().fillna(0)

# Debug: Print the grouped and summed DataFrame
print("\nAge Group Totals:")
print(age_category_totals)

# Plotting
age_category_totals.plot(kind='bar', stacked=True, figsize=(12, 8), color=['skyblue', 'orange', 'green'])
plt.title('Total Purchases by Age Group and Category')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.show()
```

```python
cust_data_columns=cust_data.columns
cust_data_columns
```

```python
def explore_object_type(df ,feature_name):  
    if df[feature_name].dtype ==  'object':
        print(df[feature_name].value_counts())
```

```python
explore_object_type(cust_data, 'Gender')
```

```python
for featureName in cust_data_columns:
    if cust_data[featureName].dtype == 'object':
        print('\n"' + str(featureName) + '\'s" Values with count are :')
        explore_object_type(cust_data, str(featureName))
```

Encoding the Boolean types of columns

```python
cust_data.Gender = cust_data.Gender.replace({'Male' : 1, 'Female' : 0})

cust_data.Subscription = cust_data.Subscription.replace({'Yes' : 1, 'No' : 0})

cust_data.Discount = cust_data.Discount.replace({'Yes' : 1, 'No' : 0})

cust_data.Promo_Code_Used = cust_data.Promo_Code_Used.replace({'Yes' : 1, 'No' : 0})
```

```python
from sklearn.preprocessing import LabelEncoder
feature_col = ['Category','Size', 'Season', 'Payment Method', 'Shipping Type', 'Frequency of Purchases']
le = LabelEncoder()
for col in feature_col:
    cust_data[col] = le.fit_transform(cust_data[col])
   
```
